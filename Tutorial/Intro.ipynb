```python
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_full_time_US = df[(df['job_country'] == "United States") & (df['job_schedule_type'] == 'Full-time')]
```

```python
sns.barplot('df_full_time_US', x='skill_percent', y='job_skills', hue='skill_count', palette='dark:b_r')
```

```python
df_DS_US = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')]
```

```python
df_plot = df_DS_US['job_location'].value_counts().head(10).to_frame()

sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
```

```python
df_plot = df_DS_US['company_name'].value_counts().head(10).to_frame()

sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='light:b', legend=False)
sns.despine()
```
